### Función de utilidad lineal

```python


La función de utilidad lineal es una representación gráfica de una función de utilidad en economía y teoría de la decisión, donde la utilidad de un consumidor o agente es una función lineal de las cantidades de bienes o servicios que consume. En términos simples, esta curva describe cómo el nivel de satisfacción (utilidad) cambia proporcionalmente con respecto a los bienes consumidos, bajo una relación lineal.

Características principales de la función de utilidad lineal:
Relación lineal: La función de utilidad en una curva lineal tiene la forma de:


U(x,y) es la utilidad que el consumidor obtiene al consumir 

a y 𝑏 son coeficientes que reflejan la importancia o la preferencia relativa de los bienes 1 y 2.
x e y son las cantidades de los bienes consumidos.

Propiedad de la linealidad: En una funciónde utilidad lineal, la utilidad aumenta o disminuye a un ritmo constante a medida que se consumen más bienes.
Esto implica que las preferencias del consumidor son constantes y no cambian a medida que aumenta la cantidad de bienes consumidos.
Es decir, el consumidor tiene una tasa constante de sustitución entre los bienes. Por ejemplo, si la relación entre los bienes 1 y 2 es de 1 a 2, entonces el consumidor estaría dispuesto a sacrificar 2 unidades del bien 2 para obtener 1 unidad más del bien 1, y esta tasa se mantendría constante.

Curvas de indiferencia: En el caso de la utilidad lineal, las curvas de indiferencia (conjuntos de combinaciones de bienes que proporcionan el mismo nivel de satisfacción) son líneas rectas. Esto se debe a que, dada la función de utilidad lineal, cualquier cambio proporcional en los bienes sigue generando el mismo nivel de satisfacción, y no hay efectos marginales decrecientes.

Tasa de sustitución constante: La tasa a la que el consumidor está dispuesto a sustituir un bien por otro (llamada tasa marginal de sustitución) es constante. Esto se debe a que la pendiente de la curva de utilidad lineal no cambia. Matemáticamente, la tasa marginal de sustitución (
𝑀𝑅𝑆) entre dos bienes es la razón de los coeficientes 

 
Esta relación constante de sustitución es una característica clave de la utilidad lineal.

Implicaciones de la utilidad lineal:
Preferencias constantes: El consumidor tiene preferencias constantes y siempre está dispuesto a hacer sustituciones en una tasa fija entre los bienes. No hay lugar para efectos de saturación o cambios en la actitud del consumidor hacia los bienes a medida que consume más.

Falta de la ley de utilidad marginal decreciente: A diferencia de las funciones de utilidad más realistas, que muestran una utilidad marginal decreciente (es decir, cada unidad adicional de un bien genera menos satisfacción), la utilidad lineal no refleja este fenómeno. En la utilidad lineal, cada unidad adicional de un bien aumenta la utilidad en la misma cantidad que la anterior.

Simplificación matemática: La utilidad lineal es una simplificación que se utiliza a menudo en modelos económicos básicos, ya que es fácil de trabajar y analizar, aunque rara vez refleja la complejidad de las decisiones de consumo en el mundo real.

```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Función para crear el gráfico
def plot_function(a, b, ax):
    # Definir el rango de x y y
    x = np.linspace(-10, 10, 400)
    y = np.linspace(-10, 10, 400)
    X, Y = np.meshgrid(x, y)
    
    # Definir la función U(x, y) = ax + by
    Z = a * X + b * Y
    
    # Crear el gráfico
    cont = ax.contour(X, Y, Z, levels=20)
    ax.set_title(f'U(x, y) = {a}x + {b}y')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

# Crear la figura y los ejes para los tres gráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Graficar tres diferentes funciones con valores de a y b distintos
plot_function(1, 2, axs[0])
plot_function(2, -1, axs[1])
plot_function(-1, 1, axs[2])

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
