```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Funci贸n para crear el gr谩fico
def plot_function(a, b, ax):
    # Definir el rango de x y y
    x = np.linspace(-10, 10, 400)
    y = np.linspace(-10, 10, 400)
    X, Y = np.meshgrid(x, y)
    
    # Definir la funci贸n U(x, y) = ax + by
    Z = a * X + b * Y
    
    # Crear el gr谩fico
    cont = ax.contour(X, Y, Z, levels=20)
    ax.set_title(f'U(x, y) = {a}x + {b}y')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

# Crear la figura y los ejes para los tres gr谩ficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Graficar tres diferentes funciones con valores de a y b distintos
plot_function(1, 2, axs[0])
plot_function(2, -1, axs[1])
plot_function(-1, 1, axs[2])

# Mostrar el gr谩fico
plt.tight_layout()
plt.show()
```
