```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la funcin CES
def ces_utility(x, y, a, b, r):
    return (a * x**r + b * y**r)**(1/r)

# Crear un grid de puntos (x, y)
x = np.linspace(0.1, 10, 400)
y = np.linspace(0.1, 10, 400)
X, Y = np.meshgrid(x, y)

# Parmetros a, b, r
a = 1
b = 1
r = 0.5  # Puedes cambiar el valor de r para ver cmo cambia la curva

# Calcular los valores de U(x, y)
Z = ces_utility(X, Y, a, b, r)

# Graficar la superficie 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Etiquetas y ttulo
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('U(x, y)')
ax.set_title(f'Curvas CES con a={a}, b={b}, r={r}')

# Mostrar la grfica
plt.show()
```
